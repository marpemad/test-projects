```python
# loading libraries
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
import pandas as pd
import os
os.chdir('c:/users/nicolas/documents/data/electric-motor-temperature')
```

```python
# importing a random regression dataset
df = pd.read_csv('pmsm_temperature_data.csv')
df.profile_id.value_counts()
df = df.loc[df['profile_id'] == 20] 
df.drop(['profile_id'], axis=1, inplace=True)
print(df.head())
```

```python
# plot settings
plot = figure(plot_width=400, plot_height=400, tools='pan,box_zoom', x_axis_label='Stator Yoke',
              y_axis_label='Stator Tooth') 
```

```python
# making the plot
plot.circle(df['stator_yoke'], df['stator_tooth'])
output_notebook()
show(plot)
```
